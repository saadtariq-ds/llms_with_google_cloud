## Predictions with Context

```python
def summary(company, crunchbase_url, city, country, industry, investor_list):
    investors = "The investors in the company are"
    for investor in ast.literal_eval(investor_list):
        investors += f"{investor}, "

    text = f"{company} has headquarters in {city} in {country} and is in the field of {industry}. {investors}. You can find more information at {crunchbase_url}"

    return text
```

```python
df['summary'] = df.apply(lambda df: summary(df['Company'], df['Crunchbase Url'], df['City'], df['Country'], df['Industry'], df['Investors']), axis=1)
```

### Creating Embeddings of Sample DataFrame

```python
sample_df = df.head(5)
```

```python
len(sample_df)
```

```python
def get_summmary_embeddings(summary):
    return embedder.get_embeddings([summary])[0].values
```

```python
sample_df['embeddings'] = sample_df['summary'].apply(get_summmary_embeddings)
```

```python
sample_df.head()
```

### Similarity Search

```python
def vector_similarity(A, B):
    vec1 = np.array(A)
    vec2 = np.array(B)

    return np.dot(np.array(vec1), np.array(vec2))
```

```python
vector_similarity(sample_df.iloc[0]['embeddings'], sample_df.iloc[1]['embeddings'])
```

#### Similarity Search from a Query

```python
def embed_prompt_lookup():
    question = input("What question do you have about a startup?")

    # Embebbings
    prompt_embedding = get_summmary_embeddings(question)

    # Similarity Search
    sample_df['prompt_similarity'] = sample_df['embeddings'].apply(lambda vector: vector_similarity(vector, prompt_embedding))

    # Most Similar Embebbing
    context = sample_df.nlargest(1, 'prompt_similarity').iloc[0]['summary']

    # Prompt + Context
    prompt = f""" Only answer the question below if you have 100% certainty of the facts. Use the context below to answer
    Here is some context:
    {context}

    Question: {question}
    """

    result = model.predict(prompt=prompt, temperature=0, max_output_tokens=2048)
    print(result.text)
```

```python
embed_prompt_lookup()
```
